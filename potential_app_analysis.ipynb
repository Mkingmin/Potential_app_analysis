# Data cleaning
## Delete wrong data
The Google Play data set has a dedicated discussion [section](https://www.kaggle.com/datasets/lava18/google-play-store-apps/discussion?sort=hotness), and we can see that one of the discussions outlines an [error](https://www.kaggle.com/datasets/lava18/google-play-store-apps/discussion/164101) for row 10472.

The problem of data in that row is the mismatching between header and value in Category and Rating fields. Because the maxinimun rating score in Google Play is 5, but data is 19. ## Check and delete duplicate values

### Check and delete duplicate values in Google Play dataset

```python
#check the total number of duplicate values
duplicate=df_gg.duplicated(subset= ["App","Category"], keep=False)
total_dup= duplicate.sum()
total_dup
```

Duplicate apps are defined as apps that have the same app name and category. There are 1847 duplicate values are found in Google Play dataset.

To make sure that the dataset show the meaningful results for analysis, delete duplicate apps that have lower rating score and keep the apps with higher rating score

```python
#sort dataset by App, Category, and Rating
df_gg=df_gg.sort_values(by=["App", "Category","Rating"], ascending=[True, True, False])

#delete duplicate values
df_gg= df_gg.drop_duplicates(subset=["App", "Category"], keep="first")
df_gg.shape[0]
```

The Google Play dataset remains 9744 rows after deleting duplicate values

### Check and delete duplicate values in App store dataset

```python
# check the total number of duplicate values
duplicate= df_as.duplicated(subset=["track_name", "prime_genre"], keep=False)
total_dup=duplicate.sum()
total_dup
```

```python
# delete duplicate values
df_as= df_as.drop_duplicates(subset=["track_name", "prime_genre"], keep="first")
print("Total rows of App Store data:", df_as.shape[0])
```

## Check and remove Non-English Apps

Because the Non-English apps which is defined as the ones that have non-English name aim at the different target audiences from whom we target to offer. We will check whether the dataset includes non-English apps. If there are some non-English apps, remove them

### Check and remove Non-English apps on Google Play dataset

```python
# Use regular expressions library to check whether app name is English

# Define function
def check_non_english_app(app_name):
    # Regex to match characters outside a-z, A-Z, 0-9, and basic punctuation
    return len(re.findall(r'[^\x00-\x7F]', app_name))

# Apply the check_non_english_app function
non_english_app=df_gg[df_gg["App"].apply(check_non_english_app) >=3]
print("Total Non-English apps:", non_english_app.shape[0])
```

Using the Regular expressions library to define a function that runs through each line of data to check if the app name is in English or not. To minimize the impact of data cleaning on results of analysis, I consider apps that have more than or equal to 3 non-ASCII characters as non-English apps which need removing from the Google Play dataset 

The result shows that there are 62 Non_English apps which need removing from the Google Play dataset

```python
# Remove Non-English apps
df_gg=df_gg[df_gg["App"].apply(check_non_english_app)<3]
print("Total Non-English apps:", df_gg.shape[0])
```

After removing non-English apps, wrong data, and duplicate values as well, Google Play dataset has 9682 rows that are ready to further analysis.

### Check and remove Non-English apps on App Store dataset

```python
# Check non-English apps in App Store dataset using check_non_english_app function defined above

non_english_app=df_as[df_as["track_name"].apply(check_non_english_app) >=3]
print("Total number of Non-English apps on App Store:", non_english_app.shape[0])
```

```python
# Remove non-English apps from App Store dataset
df_as=df_as[df_as["track_name"].apply(check_non_english_app) <3]
df_as.shape[0]
```

After removing non-English apps, wrong data, and duplicate values as well, App store dataset has 6153 rows that are ready to further analysis.

As mentioned in the Overview and Objective sections, the project that analysis supports is an ad-supported mobile app - a free app profiting from in-app ads. So, only data about free apps should be took into account

## Explore frequency by Genre and Category on Google Play and App Store free apps

```python
# Create frequency table by Genre of Google Play free apps
frequency_table_gg= free_apps_gg.groupby(["Genres"]).size().reset_index(name="Frequency")

# Calculate the % of frequency
frequency_table_gg["Percentage"]= (frequency_table_gg["Frequency"]/len(free_apps_gg))*100
frequency_table_gg= frequency_table_gg.sort_values(by="Percentage", ascending=False)
frequency_table_gg
```

The top 5 genres with the largest number of free apps on Google Play are Tools, Entertainment, Education, Business, and Productivity, with frequencies that are not significantly different from each other. This shows that app genres supporting users in work and life are as diverse as entertainment apps

```python
# Create frequency table by Category of Google Play free apps
frequency_cate_gg= free_apps_gg.groupby(["Category"]).size().reset_index(name="Frequency")

# Calculate the % of frequency
frequency_cate_gg["Percentage"] = (frequency_cate_gg["Frequency"] / len(free_apps_gg))*100
frequency_cate_gg= frequency_cate_gg.sort_values(by="Percentage", ascending= False)
frequency_cate_gg.head(10)
```

The difference between the **Genres** and the **Category** columns is not crystal clear, but one thing we can notice is that the Genres column is much more granular (it has more categories). The overview picture of category shows that the majority of free apps on Google Play relate to Family topics

```python
# Create frequency table by Genre of App Store free apps
frequency_table_as= free_apps_as.groupby(by="prime_genre").size().reset_index(name="Frequency")

# Calculate percentage
frequency_table_as["Percentage"] = (frequency_table_as["Frequency"] / len(free_apps_as))*100
frequency_table_as=frequency_table_as.sort_values(by="Percentage", ascending=False)
frequency_table_as.head(10)
```

Most free apps on the App Store are Games, accounting for 58% of all free apps. This reveals a significant imbalance in the distribution of free app genres on the App Store, where for-fun apps is the dominant genre.

## Evaluate users interest through downloads by Genre on Google Play

Use **Installs** field to analyze customers interest by Genre

```python
# Check data type of Installs field
free_apps_gg["Installs"].head(5)
```

```python
# Remove the "+" at the end of values
free_apps_gg["Installs"]=free_apps_gg["Installs"].str.rstrip("+")

# Remove the "," character
free_apps_gg["Installs"]=free_apps_gg["Installs"].str.replace(",","", regex=False)

# Convert Installs field into float
free_apps_gg["Installs"]=free_apps_gg["Installs"].astype(float)
```

```python
# Calculate the average number of downloads by Genre
avg_downloads=free_apps_gg.groupby(by="Genres")["Installs"].mean().reset_index(name="Avg_num_downloads").sort_values(by="Avg_num_downloads", ascending=False)
avg_downloads.head(30)
```

The top 5 genres with the highest download numbers on Google Play are Adventure, Communication, Casual, Video Players & Editors, and Social. However, it's essential to examine whether a few outlier apps are dominating the download numbers, potentially distorting the overall picture.

```python
# Calculate the average number of downloads by Genre in FAMILY category
family= free_apps_gg[free_apps_gg["Category"]=="FAMILY"]
family_num= family.groupby(by="Genres")["Installs"].mean().reset_index(name="Total_downloads").sort_values(by="Total_downloads", ascending=False)
family_num.head(10)
```

This suggests that Casual games have a strong appeal to users, making it a genre worth further exploration.

## Evaluate users interest through downloads by Genre on App Store
One way to find out what genres are the most popular (have the most users) is to calculate the average number of installs for each app genre. For the Google Play data set, we can find this information in the Installs column, but for the App Store data set this information is missing. As a workaround, we'll take the total number of user ratings as a proxy, which we can find in the rating_count_tot app.

```python
# Calculate the average number of rating by Genre on App Store
avg_rating=free_apps_as.groupby(["prime_genre"])["rating_count_tot"].mean().reset_index(name="Avg_num_rating").sort_values(by="Avg_num_rating", ascending=False)
avg_rating.head(10)
```

```python
# Calculate the average number of rating by App of each Genre
# rating by App on Navigation genre
navigate= free_apps_as[free_apps_as["prime_genre"]=="Navigation"]
navigate_rating = navigate.groupby(["track_name"])["rating_count_tot"].mean().reset_index(name="Avg_num_rating").sort_values(by="Avg_num_rating", ascending=False)

# rating by App on Reference genre
reference= free_apps_as[free_apps_as["prime_genre"]=="Reference"]
reference_rating = reference.groupby(["track_name"])["rating_count_tot"].mean().reset_index(name="Avg_num_rating").sort_values(by="Avg_num_rating", ascending=False)

# rating by App on Socical Networking
social= free_apps_as[free_apps_as["prime_genre"]=="Social Networking"]
social_rating = social.groupby(["track_name"])["rating_count_tot"].mean().reset_index(name="Avg_num_rating").sort_values(by="Avg_num_rating", ascending=False)

# rating by App on Music
music= free_apps_as[free_apps_as["prime_genre"]=="Music"]
music_rating = music.groupby(["track_name"])["rating_count_tot"].mean().reset_index(name="Avg_num_rating").sort_values(by="Avg_num_rating", ascending=False)

# rating by App on Book
book= free_apps_as[free_apps_as["prime_genre"]=="Book"]
book_rating = book.groupby(["track_name"])["rating_count_tot"].mean().reset_index(name="Avg_num_rating").sort_values(by="Avg_num_rating", ascending=False)
```
